```python
def outliers_detection(data,columnName):
    q1 = data[columnName].quantile(0.25)
    q3 = data[columnName].quantile(0.75)
    iqr = q3-q1
    
    upper=q3+(1.5*iqr)
    lower=q1-(1.5*iqr)
    
    return lower,upper
```

```python
outliers_detection(df,'Ozone')
```
